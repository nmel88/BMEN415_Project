```python
#importing necessary packages
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#importing data sets that has already been preprocessed (null values have been dealt with)
trainData = pd.read_csv('QSARtrain.csv')
testData = pd.read_csv('QSARtest.csv')
trainData.head() #checking data upload with a visual ouput
```

```python
#seperating inputs (x's) and outputs (y's) for test train data
xtrain = trainData.drop('logBCF', axis=1)
ytrain = trainData['logBCF']
xtest = testData.drop('logBCF', axis=1)
ytest = testData['logBCF']
```

```python
#training model with training data
reg = linear_model.LinearRegression()
#reg.fit(xtrain,ytrain)
ypred = reg.fit(xtrain,ytrain).predict(xtest)
print("Coefficients: \n", reg.coef_)
```

```python
#mean squared error
print("Mean squared error: %.2f" % mean_squared_error(ytest, ypred))
#coefficient of determination (1 is perfect prediction)
print("Coefficient of determination: %.2f" % r2_score(ytest, ypred))
```
